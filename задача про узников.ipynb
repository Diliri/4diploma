# Задача: 
Двом ізольованим один від одного в'язням дають дві різні нескінченні послідовності нулів та одиниць. Перший в'язень бачить свою послідовність і каже номер позиції, на якій в нього стоїть 0 або 1. Другий в'язень бачить свою послідовність і каже цифру - номер позиції, на якій в нього стоїть теж 0 або 1. Вбивця, який тримає їх в ув'язненні, приймає їхні відповіді (номери позицій) і забирає їх послідовності. Вбивця спочатку дивиться на послідовність першого в'язня і згадує, який номер позиції назвав другий в'язень, - фіксує цифру на вищеописаній позиції в цій послідовності (послідовності першого).
Бере потім послідовність другого в'язня і згадує номер позиції, яку назвав перший в'язень. Робить те саме.
Потім вбивця порівнює цифри на цих позиціях - Якщо цифри співпадають (тобто 00 або 11), то він відпускає в'язнів, якщо цифри різні (тобто 01 або 10), то він їх убиває.
Придумати стратегію для двох ув'язнених, щоб ймовірність вижити у них була строго більше ніж 1/2.
# Розв'язок (одна зі стратегій):
Тут варто зробити важливе зауваження, що ця стратегія працюватиме в умовах, якщо два в'язні думають однаково, а це в житті не можна гарантувати - засобів зв'язку в'язні не мають (вони абсолютно ізольовані одне від одного та від інтернету і світу). Тому робиться припущення, що в'язні думають однаково.
Однаково, в якому плані?
Перший і другий в'язень будуть дивитися на n перших цифр.
Отже, стратегія така: назвати позицію першого нуля, а якщо нуля в послідовності перших зафіксованих числах нема, то назвати довжину послідовності.
Наприклад, якщо і перший, і другий в'язні фіксують перші дві цифри з послідовності, то в першого, нехай буде 00, а в другого - 11. Перший називає позицію 1, другий - 2. В першого на позиції 2 стоїть цифра «0», а в другого на позиції 1 стоїть цифра «1», тоді вбивця їх убиває.
Робимо зауваження, що фіксувати перші цифри варто, починаючи з n=2, бо при n=1 ймовірність вижити = 1/2.

```python
# сначала фиксируем к-во первых цифр: (т.е. repeat = n от двух и больше).
# количество размещений с повторениями (англ. permutations with repetition ) 
n = 3
import itertools
x = [0, 1] 
k = [p for p in itertools.product(x, repeat = n)]
combinations = [p for p in itertools.product(k, repeat=2)]
print(combinations)
#print(combinations[0][1])
#print(combinations[0][1][1])

# cтратегія - номер першого нуля, а якщо нуля нема, то кажуть довжину перших цифp

res = []
for i in combinations:
    listed_data = []
    for j in i:
# Finding the index of the first occurrence of any item (в даному випадку - нуля) in a list              
        if 0 in j: 
            first_occurrence = j.index(0)
            listed_data.append(first_occurrence)
        else: 
            first_occurrence = (len(j) - 1)
            listed_data.append(first_occurrence)
    res.append(listed_data)
    
print(res)

something = [i for i in range(len(combinations)) if combinations[i][1][res[i][0]] == combinations[i][0][res[i][1]]]
#print('\nSomething: {} \n'.format(something))
print('Chains: {} \n'.format([combinations[i] for i in something]))
print('Answers: {}'.format([res[i] for i in something]))

# для того, щоб порахувати ймовірність, визначаємо простір Омега :
print('\n', len(combinations))
print('\n', len(something), '\n')

prob = len(something) / len(combinations)
print(prob)
```
