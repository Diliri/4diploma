```python
# https://datatofish.com/read_excel/

import pandas as pd

data = pd.read_excel (r'C:\Users\Diana\Downloads\Telegram Desktop\time2.xlsx')
```

```python
data
```

```python
Lst = list(data.columns.values)
print('Type: {} \n'.format(type(Lst)) ) # list

new_Lst = []
for el in Lst:
    # https://www.programiz.com/python-programming/methods/string/startswith
    if el.startswith('Прізвище'):
        new_Lst.append(el)
print('Респонденти, яких треба буде фільтрувати, знаходяться в стовпцях під назвами: {} \n'.format(new_Lst) )

df = pd.DataFrame(data, columns = new_Lst)
print ('Респонденти: \n', df)
```
