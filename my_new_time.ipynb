```python
Dct = dict()

def to_sec(time, word):
    if time.find(word) != -1:
        index = time.find(word)
        sec = int(time[:index])
        new_time = time[index + len(word):]
        return sec, new_time
    return 0, ''

def parse_time(time):
    total = 0
    words = ['днів', 'година', 'години', 'хв', 'сек']
    words_tmp = ['дні', 'годин', 'х', 'се']
    for word in words:
        sec, new_time = to_sec(time, word)
        word_index = words_tmp.index(word[:-1])
        if word_index != 0:
            total += sec * 60**(len(words_tmp) - word_index - 1)
        else:
            total += sec * 24 * 60**2
        if new_time: time = new_time
    return total

for col in data:
    if not col.startswith('Прізвище'):
        continue
    for i, el in enumerate(data[col]):
        try:
            if int(el) not in Dct.keys():
                Dct[int(el)] = {'count': 1}
                if col[-1].isnumeric():
                    Dct[int(el)]['time'] = [data['Затрачений час.' + col[-1]][i]]
                else:
                    Dct[int(el)]['time'] = [data['Затрачений час'][i]]
            else:
                Dct[int(el)]['count'] += 1
                if col[-1].isnumeric():
                    Dct[int(el)]['time'].append(data['Затрачений час.' + col[-1]][i])
                else:
                    Dct[int(el)]['time'].append(data['Затрачений час'][i])
        except:
            continue
            
for k in Dct.keys():
    for i in range(len(Dct[k]['time'])):
        Dct[k]['time'][i] = parse_time(Dct[k]['time'][i])
    
print(Dct)
```
